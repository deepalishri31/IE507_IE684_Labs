Ex 1.7 \\
Implement your procedure for the data from S1.txt and report the best choice for the number of clusters. 


```python
distortions = []
K = range(1,20)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(X)
    distortions.append(kmeanModel.inertia_)

plt.plot(K, distortions, 'bx-')
plt.xlabel('Num. of clusters')
plt.ylabel('Distortion')
plt.show()
```

Ex 1.15 \\
Using your procedure to find the best choice of number of clusters, report the best choice for the number of clusters for the column normalized data. \\


```python
distortions = []
K = range(1,20)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(NX)
    distortions.append(kmeanModel.inertia_)

plt.plot(K, distortions, 'bx-')
plt.xlabel('Num. of clusters')
plt.ylabel('Distortion')
plt.show()
```
